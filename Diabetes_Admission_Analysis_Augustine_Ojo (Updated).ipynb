## **Observations And Inferences On Classification Report:**
Classification report is the summary of the performance of a classification model to evaluate its performance:
* *Precision:*  The ratio of correctly predicted positive observation.
* *Recall/Sensitive:*  The ratio of correctly predicted positive observations.
* *F1-Score:*  The harmonic mean of precision and recall.
* *Support:*  The actual number of occurences of the class in the dataset.
* *Macro Accuracy:*  The average of the calculated accuracy of each individual class, where each class contribute equally, regardless of count imbalances.
* *Weighted Accuracy:*  The weighted average of the accuracy score based on the number of samples in each class. It favours class with larger number.

The Macro Accuracy is equal to the Weighted Accuracy at **0.78.** This could be due to the fact that same number **54,649 ** was used used to calculate both. more investigation is required for the iterpretation of this result.
